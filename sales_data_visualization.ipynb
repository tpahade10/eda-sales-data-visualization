# **Data Analysis and Visualization**

```python
fig , ax=plt.subplots(figsize=(8,5))
rev_by_contry=df.groupby(['COUNTRY'])['SALES'].sum()
rev_by_contry.plot.bar()
plt.ticklabel_format(style='plain', axis='y')
plt.title('Sales by Country',fontdict={'fontsize':15})
plt.show()

```

```python
import calendar

# Assuming 'ORDERDATE' is a string column representing the date in the format 'dd/mm/yyyy'
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'], format='%d/%m/%Y')

# Extract month and year from the ORDERDATE column
df['Month'] = df['ORDERDATE'].dt.month
df['Year'] = df['ORDERDATE'].dt.year

# Group by month and year, sum the QUANTITYORDERED
monthly_quantity = df.groupby(['Year', 'Month'])['SALES'].sum()

# Print all months and quantities, grouped by year
current_year = None
for (year, month), quantity in monthly_quantity.items():
    if year != current_year:
        print(f"\nYear {year}")  # Print year header once for each year
        current_year = year
    month_abbr = calendar.month_abbr[month]  # Convert month number to abbreviation
    print(f"{month_abbr}: {quantity}")
```

```python
fig , ax=plt.subplots(figsize=(8,5))
rev_by_contry=df.groupby(['PRODUCTLINE'])['SALES'].sum()
rev_by_contry.plot.bar()
plt.ticklabel_format(style='plain', axis='y')
plt.title('Profit by Product',fontdict={'fontsize':15})
plt.show()

```
